```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# Loading the dataset 
iris_dataset = pd.read_csv("IRIS.csv") 
iris_dataset.head()
```

- We can see that our data has no missing values 

```python
# Looking at the relationship between variables 
# Relationship between sepal length and sepal width 

sns.scatterplot(iris_dataset, x="sepal_length", y="sepal_width", hue="species", palette={"Iris-setosa":'r', "Iris-versicolor":'g', "Iris-virginica":"b"})

# Setting the legend outside the figure 
plt.legend(bbox_to_anchor=(1,1), loc=2)
plt.title("Relationship B2N Sepal Width and Sepal Length")
```

From the observation;
- Sepal width for Iris Setosa is longer that sepal length. 
- For most of the Iris Versicolor, sepal width and sepal length are almost equal in size. 
- Iris Virginica has larger sepal length that sepal width.

```python
# Looking at the relationship between petal length and petal width 
sns.scatterplot(iris_dataset, x="petal_width", y="petal_length", hue="species", palette={"Iris-setosa":'r', "Iris-versicolor":'g', "Iris-virginica":"b"})
```

From the plot;
- Iris setosa has smaller petal width and petal length.
- Iris virginica has larger petal lengths and petal widths. 
- Iris versicolor lies at the middle of setosa and virginica 

### Looking at the relationship between columns

```python
sns.pairplot(iris_dataset.loc[:, "sepal_length": "species"], hue="species", palette={"Iris-setosa":'r', "Iris-versicolor":'g', "Iris-virginica":"b"})
```

```python
# Comparing petal widths accros species 
fig, ax = plt.subplots(1,2, figsize=(12, 6), sharex=True)

sns.boxplot(iris_dataset, x="species", y="petal_width", ax=ax[0])
sns.boxplot(iris_dataset, x="species", y="petal_length", ax=ax[1])

ax[0].set_title("Compare Petal Width Accross Species")
ax[1].set_title("Compare Petal Length Accross Species")
```

Observations 
- Setosa has smaller petal width and petal length compared to the rest of the species.
- Virginica has large petal width and petal length compared to all other species.

```python
# Comparing petal sepal lengths accross species 
fig, ax = plt.subplots(1, 2, figsize=(12, 6)) 

sns.boxplot(iris_dataset, x="species", y="petal_length", ax=ax[0])
ax[0].set_title("Comparing Sepal Length Across Species") 

sns.boxplot(iris_dataset, x="species", y="sepal_width", ax=ax[1]) 
ax[1].set_title("Comparing Sepal Width Across Species")
```

Observations 
- Iris Setosa has the smallest petal length but the largest petal width. 
- Iris Virginica has the largest petal length but falls second sepal width. 
- Versicolor has the smallest petal width.
